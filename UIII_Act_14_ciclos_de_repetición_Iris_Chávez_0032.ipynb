#### Problema 11: Escribe un programa que diga si un número introducido por teclado es o no primo. Un número primo es aquel que sólo es divisible entre él mismo y la unidad. Nota: Es suficiente probar hasta la raíz cuadrada del número para ver si es divisible por algún otro número.


```python
# --- PROGRAMA INTERACTIVO PARA VERIFICAR NÚMEROS PRIMOS ---

print("Problema 11: Determinación de números primos.")
print("Un número primo es aquel que solo es divisible por sí mismo y la unidad.")

# Entrada de datos interactiva por teclado convirtiendo a entero
numero = int(input("Ingrese un número entero positivo: ")) # [3]

# Validación inicial: Los números menores o iguales a 1 no son primos [1]
if numero <= 1: # [4, 5]
    print(f"El número {numero} no es un número primo.") # [6]
else:
    # Se utiliza una variable lógica como bandera para el estado del número [2, 7]
    es_primo = True

    # Se calcula la raíz cuadrada del número para limitar las iteraciones [1, 8, 9]
    # En Python, el operador ** se utiliza para potencias [8]
    limite = int(numero**0.5)

    # Se inicializa el divisor en 2 para comenzar la verificación [10]
    divisor = 2

    # Ciclo while que se repite mientras el divisor sea menor o igual al límite [2, 11]
    while divisor <= limite: # [5]
        # Se utiliza el operador de módulo % para obtener el resto de la división [8]
        if numero % divisor == 0: # [4]
            # Si el residuo es cero, tiene un divisor adicional y no es primo [1, 2]
            es_primo = False
            # La sentencia break permite salir del bucle inmediatamente [12]
            break

        # Incremento del divisor en cada iteración del ciclo [10, 12]
        divisor = divisor + 1

    # Estructura selectiva para imprimir el diagnóstico final [3, 4]
    if es_primo:
        print(f"Resultado: El número {numero} es un número primo.") # [6, 13]
    else:
        print(f"Resultado: El número {numero} no es un número primo.") # [3, 13]
    print("*/*/*/*/*//*/*/*/*/*/*/*/*/*/*/**")
    print("Programa realizado por Iris Chavez NL = 11")
```
